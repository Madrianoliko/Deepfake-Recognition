```python
import cv2
import tensorflow as tf
import numpy as np

CATEGORIES = ['real', 'fake']
def prepare(filepath):
    IMG_SIZE = 100
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE) 
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)

model = tf.keras.models.load_model('mesonet_deepfake_recognition.model')

tested_img_path1 = r"C:\Users\adria\Desktop\Praca Magisterska\kaggle-challenge\data_to_train\real\1457.jpg"
tested_img_path2 = r"C:\Users\adria\Desktop\Praca Magisterska\kaggle-challenge\data_to_train\real\343.jpg"
tested_img_path3 = r"C:\Users\adria\Desktop\Praca Magisterska\kaggle-challenge\data_to_train\real\656.jpg"
```

```python
prediction = model.predict([prepare(tested_img_path1)])
print(CATEGORIES[int(prediction[0][0])])
print(prediction[0][0])
```

```python
prediction = model.predict([prepare(tested_img_path2)])
print(CATEGORIES[int(prediction[0][0])])
print(prediction[0][0])
```
