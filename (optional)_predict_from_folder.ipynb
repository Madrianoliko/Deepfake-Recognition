```python
import os
import cv2
import tensorflow as tf
import numpy as np

CATEGORIES = ['real', 'fake']
fake_dir_path = r"C:\Users\adria\Desktop\Praca Magisterska\kaggle-challenge\data_to_test\fake"
real_dir_path = r"C:\Users\adria\Desktop\Praca Magisterska\kaggle-challenge\data_to_test\real"
model = tf.keras.models.load_model('mesonet_deepfake_recognition.model')

def create_predict_array_from_dir(filepath_dir):
    array_to_predict = []
    img_dir = os.listdir(filepath_dir)
    
    for img_path in img_dir:
        path_img = os.path.join(filepath_dir, img_path)
        IMG_SIZE = 100
        img_array = cv2.imread(path_img, cv2.IMREAD_GRAYSCALE) 
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        array_to_predict.append(new_array.reshape(IMG_SIZE, IMG_SIZE, 1))
    return np.array(array_to_predict)
```

```python
def get_prediction_result_from_real_folder(predictions):
    TP = 0
    FN = 0
    for prediction in predictions:
        if CATEGORIES[int(prediction)] == "real":
            TP += 1
        else:
            FN += 1

    return (TP, FN)
    
```

```python
def get_prediction_result_from_fake_folder(predictions):
    FP = 0
    TN = 0
    for prediction in predictions:
        if CATEGORIES[int(prediction)] == "real":
            FP += 1
        else:
            TN += 1

    return (FP, TN)
```

```python
array_to_predict = create_predict_array_from_dir(real_dir_path)
print(array_to_predict.shape)
predictions = model.predict(array_to_predict)
TP, FN = get_prediction_result_from_real_folder(predictions)
```

```python
array_to_predict = create_predict_array_from_dir(fake_dir_path)
print(array_to_predict.shape)
predictions = model.predict(array_to_predict)
FP, TN = get_prediction_result_from_fake_folder(predictions)
```

```python
print("Zdiagnozowane jako prawdziwe (True posititive): {}".format(TP))
print("Zdiagnozowano jako fałszywe (False Negative): {}".format(FN))
print("Zdiagnozowane jako fałszywe (True Negative): {}".format(TN))
print("Zdiagnozowano jako prawdziwe (False Positive): {}".format(FP))
```

```python
print("Accuracy: {}".format((TP+TN)/(TN+TP+FN+FP)))
print("Precision: {}".format(TP/(TP+FP)))
```
